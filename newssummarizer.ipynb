```python
from transformers import pipeline
summarizer = pipeline("summarization", model="Falconsai/text_summarization")
text = """Express Investigation: One nation, a few parivars. In Bihar, RJD leads family front
Congress’s political footprint may have shrunk but its share of dynast-legislators (MPs, MLAs and MLCs) is at 33.25%. With almost thrice the number of legislators, BJP’s share is 18.62%; families dominate regional politics; 149 families across parties have more than one dynast-legislator. """
```

```python
summary = summarizer(text, max_length=50, min_length=20, do_sample = True)
print("Original Text:\n",text)
print("\nSummary: \n", summary[0]['summary_text'])
```

```python
%%writefile app.py
from flask import Flask, request, render_template
from transformers import pipeline
from newspaper import Article

# ✅ Create Flask app
app = Flask(__name__)

# ✅ Initialize BART summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# ✅ Function to summarize text or article URL
def summarize_text_or_url(input_text=None, input_url=None):
    """
    Summarizes either direct text input or article content from a given URL.
    """
    if input_url:
        try:
            article = Article(input_url)
            article.download()
            article.parse()
            text = article.text
        except Exception as e:
            return f"❌ Error fetching article: {e}"
    elif input_text:
        text = input_text
    else:
        return "⚠️ Please provide text or a URL to summarize."

    if not text.strip():
        return "⚠️ No valid text found to summarize."

    # Truncate text for model input (limit to avoid overload)
    text = text[:4000]

    # Generate summary
    summarized = summarizer(text, max_length=200, min_length=50, do_sample=False)
    return summarized[0]['summary_text']

# ✅ Home route (form UI)
@app.route("/", methods=["GET", "POST"])
def home():
    summary = None
    if request.method == "POST":
        input_text = request.form.get("input_text")
        input_url = request.form.get("input_url")
        summary = summarize_text_or_url(input_text, input_url)

    return render_template("index.html", summary=summary)

# ✅ Run the app on port 8000
if __name__ == "__main__":
    app.run(port=8000)
```
